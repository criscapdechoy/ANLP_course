```python
with open("resources/DrugBank.txt","r") as fp:
    lines = fp.readlines()
ents = [line.replace("\n","").split("|") for line in lines]
ents
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(ents, columns=["token","type"])
df
```

```python
df.groupby("type").size()
```

```python
df[df.type == "group"]
```

```python
df[df["token"].str.contains("acid")]
```

```python
df[(df["token"].str.contains("agents"))]
```

```python
df[(df["token"].str.match("\d+-[A-Z]+")) & (df.type == "drug")]["token"].unique()
```
