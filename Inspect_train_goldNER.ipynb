```python
with open("resources/train_goldNER.txt","r") as fp:
    lines = fp.readlines()
ents = [line.replace("\n","").split("|") for line in lines]
ents
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(ents, columns=["sentence","offset","token", "type"])
print(df.shape)
df
```

```python
df.groupby("type").size()
```

```python
drugs = df[df["type"] == "drug"]
drugs
```

```python
drugs_n = df[df["type"] == "drug_n"]
drugs_n
```

```python
brands = df[df["type"] == "brand"]
brands
```

```python
groups = df[df["type"] == "group"]
groups
```

```python
a = df[df["token"].str.match(r"[A-Z]{2,4}")]
print(a.groupby("type").size())
a.head(5) #omega-agatoxin
```

```python
a = df[df["token"].str.endswith("kers")]
print(a.groupby("type").size())
a.head(5)
```

```python
a = df[(df["token"].str.contains("-"))].copy()
a["hyphens"] = a["token"].apply(lambda x: sum("-" == e for e in x))
b = a[a["type"] == "drug_n"]
print(b.groupby("hyphens").size())
a.head(5)
```

```python
a = df[(df["token"].str.contains("toxin"))].copy()
print(a.groupby("type").size())
a.head(5)
```

```python
a = df[df["token"].str.endswith("navir")]
print(a.groupby("type").size())
a.drop_duplicates(subset="token")
```

```python
a = df[(df["type"] == "group") & (~df["token"].str.contains(" "))].copy()
a["suf4"] = [e[-4:] for e in a["token"].values]
a[["suf4"]].groupby("suf4").size().sort_values(ascending=False)[:30]
```

```python
a = df.copy()
a["len"] = a["token"].apply(len)
a["capitals"] = a["token"].apply(lambda x: sum(e.isupper() for e in x))
a["capital"] = a["token"].apply(lambda x: x.isupper())
a["lower"] = a["token"].apply(lambda x: x.islower())
a[a["type"] == "drug"]
```
